<h1>Multiples Lines, Single Plot</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
```

<h2>random low and high temperature data</h2>

```python
data_set_size = 15
low_mu, low_sigma = 50, 4.3
low_data_set = low_mu + low_sigma * np.random.randn(data_set_size)
high_mu, high_sigma = 57, 5.2
high_data_set = high_mu + high_sigma * np.random.randn(data_set_size)

days = list(range(1, data_set_size + 1))
```

```python
days
```

```python
plt.plot(days, low_data_set)
plt.show()
```

```python
plt.plot(days, low_data_set,
         days, high_data_set)
plt.show()
```

```python
plt.plot(days, low_data_set,
         days, low_data_set, "vm",
         days, high_data_set,
         days, high_data_set, "^k")
plt.show()
```

```python
plt.plot(days, low_data_set,
         days, high_data_set
         )
plt.xlabel('Day')
plt.ylabel('Temperature: degrees Fahrenheit')
plt.title('Randomized temperature data')
plt.show()
```

```python
plt.plot(days, high_data_set, "^k")
plt.xlabel('Day')
plt.ylabel('Temperature: degrees Fahrenheit')
plt.title('Randomized temperature data')
plt.show()
```

<h2>pylab_examples example code: legend_demo2.py</h2>
<h3>[matplotlib.org] (http://matplotlib.org/examples/pylab_examples/legend_demo2.html)</h3>

```python
t1 = np.arange(0.0, 2.0, 0.1)
t2 = np.arange(0.0, 2.0, 0.01)

# The plot returns a list of lines.
# The "l1, = plot" usage extracts the first element of the list using tuple unpacking.
# Therefore, l1 is a line2D instance and not a sequence of lines.

l1 = plt.plot(t2, np.exp(-t2))
l1
```

```python
t1 = np.arange(0.0, 2.0, 0.1)
t2 = np.arange(0.0, 2.0, 0.01)

# The plot returns a list of lines.
# The "l1, = plot" usage extracts the first element of the list using tuple unpacking.
# Therefore, l1 is a line2D instance and not a sequence of lines.

l1, = plt.plot(t2, np.exp(-t2))
l2, l3 = plt.plot(t2, np.sin(2 * np.pi * t2), '--go', t1, np.log(1 + t1), '.')
l4, = plt.plot(t2, np.exp(-t2) * np.sin(2 * np.pi * t2), 'rs-.')

plt.legend((l2, l4), ('oscillatory', 'damped'), loc='upper right', shadow=True)
plt.xlabel('time')
plt.ylabel('volts')
plt.title('damped oscillation')
plt.show()
```
